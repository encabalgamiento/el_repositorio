# # Задание №1

Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
def fun_lemma(old_file, new_file):
    
    f = open(old_file, 'r', encoding='utf-8')
    text = f.read()
             
    from pymystem3 import Mystem
    m = Mystem()
    lemmas = m.lemmatize(text)
             
    f_2 = open(new_file, 'w', encoding='utf-8')
    f_2.write(''.join(lemmas))
             
fun_lemma("C:\\Users\\Sveta\\Desktop\\imitation\\el_repositorio\\anecdotes.txt", "C:\\Users\\Sveta\\Desktop\\imitation\\el_repositorio\\lemma_anecdotes.txt")
```

# # Задание №2

Очистите лемматизированный текст от стоп-слов.

```python
antes_file = "C:\\Users\\Sveta\\Desktop\\imitation\\el_repositorio\\lemma_anecdotes.txt"
despues_file = "C:\\Users\\Sveta\\Desktop\\imitation\\el_repositorio\\anecdotes_no_stop.txt"

import nltk
nltk.download()

f = open(antes_file, 'r', encoding='utf-8')
text = f.read()

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

sw = stopwords.words('russian')
pun = [",", ".", "``", "''", "!", "?", "...", "(", ")", "-", ":", ";"]
filtered = [w for w in word_tokenize(text) if ((w not in sw) and (w not in pun))]

f_2 = open(despues_file, 'w', encoding='utf-8')
f_2.write(' '.join(filtered))
f_2.close()
```

Посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.

```python
from collections import Counter
f_2 = open(despues_file, 'r', encoding='utf-8')
text = f_2.read()
counts = Counter(text.split())

print(counts)
print()
print(counts.most_common(20))
f_2.close()
```

Сделайте полный морфологический разбор исходного текста.

```python
from pymystem3 import Mystem
m = Mystem()

# Я сделала короткую версию файла, чтобы можно было проверить, насколько код работает
# То же самое можно сделать и в полном файле, но ждать не очень хотелось. 

despues_file = "C:\\Users\\Sveta\\Desktop\\imitation\\el_repositorio\\anecdotes_no_stop_short.txt"

morfologia = []

f_3 = open(despues_file, 'r', encoding='utf-8')
text = f_3.read()
words = text.split()
for word in words:
    try:
        res = m.analyze(word)[0]
        morfologia.append(res)
    except KeyError:
        pass

print(morfologia)
```
