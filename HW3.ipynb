## Задание №2

Очистите лемматизированный текст от стоп-слов.

```python
# Создаю для удобства переменные для лемматизированного файла со стоп словами и будущего лемматизированного файла без стоп-слов.
antes_file = "C:\\Users\\Sveta\\Desktop\\imitation\\el_repositorio\\lemma_anecdotes.txt"        
despues_file = "C:\\Users\\Sveta\\Desktop\\imitation\\el_repositorio\\anecdotes_no_stop.txt"

import nltk          # Импортирую библиотеку
nltk.download()      # Запускаю её скачивание непосредственно из программы

f = open(antes_file, 'r', encoding='utf-8')  # Открываю "старый" файл для чтения
text = f.read()                              # Читаю "старый" файл, записываю в переменную text 

from nltk.tokenize import word_tokenize      # Импортирую корпус для токенизации из библиотеки NLTK
from nltk.corpus import stopwords            # Импортирую стоп-слова для токенизации из библиотеки NLTK        

sw = stopwords.words('russian')                                         # Оставляю в списке стоп-слов только русские
pun = [",", ".", "``", "''", "!", "?", "...", "(", ")", "-", ":", ";"]  # Создаю список с возможной пунктуацией
                                                                        # Она мне тоже не нужна 

filtered = [w for w in word_tokenize(text) if ((w not in sw) and (w not in pun))] 
# Создаю новый список, там только слова из токенизированного текста из переменной text, которые НЕ топ-слова и НЕ пунктуация

f_2 = open(despues_file, 'w', encoding='utf-8')  # Открываю "новый" файл для записи
f_2.write(' '.join(filtered))                    # Записываю туда слова из списка filtered через пробел
f_2.close()                                      # Закрываю файл 
```

Посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.

```python
from collections import Counter                     # Импортирую из модуля Collections - Counter 
f_2 = open(despues_file, 'r', encoding='utf-8')     # Открываю токенизированный файл 
text = f_2.read()                                   # Читаю его 
counts = Counter(text.split())                      # Применяю Counter к тексту, разделённому на слова. Получаю словарь 

print(counts)                                       # Вывожу
print()                
print(counts.most_common(20))                       # Применяю метод most_common().Печатаю 20 наиболее распространённых. 
f_2.close()                                         # Закрываю 
```

## Задание №3

Сделайте полный морфологический разбор исходного текста. Пройдитесь циклом по списку с разборами, который выдал pymystem3, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение). 

```python
import re                            # Импортирую модуль re для использования регулярных выражений
from pymystem3 import Mystem         # Из пакета pymystem3 импортирую модуль Mystem и записываю его удобнее
m = Mystem()

despues_file = "C:\\Users\\Sveta\\Desktop\\imitation\\el_repositorio\\anecdotes_no_stop.txt"  # Записываю в перемунную текст 
f_3 = open(despues_file, 'r', encoding='utf-8')                                               # Открываю текст для чтения
text = f_3.read()                     # Читаю текст
analisis = m.analyze(text)            # Анализирую текст, получаю список словарей

morfologia = {}                       # Создаю пустой словарь

for pareja in analisis:                      #Создаю цикл, который проходится по словарям в списке
    try:
        form = pareja["text"]                #В каждом словаре значения для ключа "text" - теперь в переменной gram
        gram = pareja["analysis"][0]["gr"]   #В каждом словаре значения для ключа "gr" - теперь в переменной gram
        morfologia[form] = gram              #В том словаре, который я создала пустым - ключам form соответствую переменные gram
    
    except KeyError:                        # Оговариваю возможные исключения
        pass
    except IndexError:
        pass

las_letras_grandes = re.compile("[A-Z]+")   #Создаю регулярное выражение, чтобы искать части речи            

partes = {}                                 #Cоздаю пустой словарь                  

for fo, gr in morfologia.items():                      #Прохожусь циклом по ключам и значениям в словаре morfologia
   partes[fo] = las_letras_grandes.match(gr).group(0)  # В словаре partes. Ключ - это ключ из morgologia. А значение - это  
                                                       #  значение из словаря morfologia, но которое содержит регулярку
print()
print("Список частей речи")  # Вывожу
print(partes)


from collections import Counter  # Импортирую из модуля collections Counter

with open(despues_file, 'r', encoding='utf-8') as f_4: # Открываю файл и читаю его
    text = f_4.read()

#абсолютное количество 
todas_partes = Counter(partes.values())    #Считаю количество всех значений в словаре partes (это и есть части речи)

print()
print("Абсолютная частота частей речи")
print(todas_partes)


#относительное количество
partes_relativas = {}                                     #Создаю новый словарь 

for tex, gr in todas_partes.items():                      # Прохожусь циклом по всем парам ключ-значение из словаря Counter
    partes_relativas[tex] = gr/len(todas_partes)          # В мой новый словарь ключи переношу так, а значения делю на общее
                                                          # кол-во частей речи

print()
print("Относительная частота частей речи")
print(partes_relativas)
```
