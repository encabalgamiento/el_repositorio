```python
# Задание 1 
# Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную. 

text = "варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг"

palabras = text.split()

bigrammas = []

for num in range(len(palabras)-1):
    bigramma = palabras[num] + " " + palabras[num + 1]
    bigrammas.append(bigramma)
    print(palabras[num] + " " + palabras[num + 1])

print()
print(bigrammas)
```

```python
#Задание 3 


#1)Скопировать текст, сохранить его в переменную.
poem = "Удивительная кошка\n\nНесчастная кошка порезала лапу-\nСидит, и ни шагу не может ступить.\nСкорей, чтобы вылечить кошкину лапу\nВоздушные шарики надо купить!\n\nИ сразу столпился народ на дороге-\nШумит, и кричит, и на кошку глядит.\nА кошка отчасти идет по дороге,\nОтчасти по воздуху плавно летит!"

#2)Узнать длину текста в символах.
print(len(poem))
print()

#3)Очистить текст от пунктуации, используя list compehensions. На выходе должен получиться список слов.

words = [word.strip("!?.,-") for word in poem.split()]
print(words)
print()

#4)Объединить слова из получившегося списка в строку через пробел.

string = ' '.join(words)
print(string)
print()

#5)С помощью list comprehensions проверить, написано ли слово с большой буквы, и если да, то привести его к нижнему регистру. 
#На выходе должен получиться список из всех слов стихотворения, написанных с маленькой буквы 

new_words = []

for palabra in words:
    if palabra.islower():
        new_words.append(palabra)
    else: 
        new_words.append(palabra.lower())

print(new_words)
print()

#6)Узнать длину текста в словах.

print(len(words))
print()

#7)Узнать количество уникальных слов.

unique_words = []

for new_word in new_words:
    if new_word not in unique_words:
        unique_words.append(new_word)

print("В стихотворении " + str(len(unique_words)) + " уникальных слов.")
print()

#8)Распечатать последние 10 слов.

print(words[-10:])
print()

#9)Соединить полученный в пункте 3 список в строку. Слова в строке должны быть написаны через знак переноса строки.

sur_poem = '\n'.join(words)
print(sur_poem)
print()

#10)Распечатать с 20 по 30 символ этой строки (включительно).

sur_poem_2 = list(sur_poem)
print(sur_poem_2[19:29])
print()

#11)Разбить текст на биграммы, записать их в переменную и посчитать частотность каждой биграммы. 
#Биграммы должны быть в виде списка кортежей.

#new_words - это список, текст, приведённый к нижнему регистру и без пунктуации
bigrams = []

for num in range(len(new_words)-1):
    bigram = tuple()
    bigram = new_words[num] + " " + new_words[num + 1]
    bigrams.append(bigram)

for bigram in bigrams: 
    print("Найдено " + str(bigrams.count(bigram)) + " вхождение биграммы " + "'" + bigram + "'")


```
